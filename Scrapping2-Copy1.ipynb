```python
# We create this function in order to add this domain name
# 'http://www.firmendb.de/'to links that don't have it 
from urllib.parse import urljoin
def add_base_domain(link_list, base_domain='http://www.firmendb.de'):
  cleaned_list = []
  for e in link_list:
    if base_domain not in e:
      cleaned_list.append(urljoin(base_domain , e))
    else:
      cleaned_list.append(e)
  
  return cleaned_list
```

```python
company_data = []
for link in linkss:
    driver = webdriver.Edge()

    # Visit the website
    driver.get(link)

    # Get the page source code after the website has fully loaded
    html = driver.page_source

    # Close the browser
    driver.quit()

    # Parse the HTML using BeautifulSoup
    soup = BeautifulSoup(html, "html.parser")

    # Find the name and email within the div with class "adressbox"
    adressbox_div = soup.find("div", class_="adressbox")

    # Extract desired information using itemprop attributes
    name = adressbox_div.find("span", itemprop="name")
    if name:
        name = name.text.strip()
    else:
        name = ""
    address = adressbox_div.find("span", itemprop="streetAddress")
    if address:
        address = address.text.strip()
    else:
        address = ""
    postal_code = adressbox_div.find("span", itemprop="postalCode")
    if postal_code:
        postal_code = postal_code.text.strip()
    else:
        postal_code = ""
    
    addressLocality = adressbox_div.find("span", itemprop="addressLocality")
    if addressLocality:
        addressLocality = addressLocality.text.strip()
    else:
        addressLocality =  ""

    telephone = adressbox_div.find("dd", itemprop="telephone")
    if telephone:
        telephone = telephone.text.strip()
    else:
        telephone =  ""

    fax_number = adressbox_div.find("dd", itemprop="faxNumber")
    if fax_number:
        fax_number = fax_number.text.strip()
    else:
        fax_number = ""
   
    email_element = adressbox_div.find("dd", itemprop="email")
    if email_element:
        email = email_element.find('a')['href']
    else:
        email = ""
#     email = adressbox_div.find("dd", itemprop="email").find("a").text.strip()
    internet = adressbox_div.find("a", target="_blank")
    if internet:
        internet = internet["href"]
    else:
        internet = ""
    
    
    
        # Parse the HTML using BeautifulSoup
    soup = BeautifulSoup(html, "html.parser")
    # Find the name and email within the div with class "adressbox"
    content_div = soup.find("div", id="tab-2")

    # Extract desired information using itemprop attributes

    code = content_div.find_all("dd")
    if len(code) >= 1:
        
        if code[0]:
            ofizieller = code[0].text
        else:
            ofizieller = ""
    else:
        ofizieller = ""
    
    if len(code) >= 2:
        if code[1]:
            hauptsitz = code[1].text
        else:
            hauptsitz = ""
    else:
            hauptsitz = ""
    if code[2]:
        branche = code[2].find('a')['href']
    else:
        branche = ""
    if code[3]:
        unternehmensgegenstand = code[3].text
    else:
        unternehmensgegenstand = ""
    
    if len(code) >= 5:
        if code[4]:
            handelsregister = code[4].text
        else:
            handelsregister = ""
    else:
            handelsregister = ""
     
    if len(code) >= 6:
        if code[5]:
            rechtsform = code[5].text
        else:
            rechtsform = ""
    else:
            rechtsform = ""
            
    if len(code) >= 7:
        if code[6]:
            firmengründung = code[6].text
        else:
            firmengründung = ""
    else:
            firmengründung = ""
            
    if len(code) >= 8:
        if code[7]:
            stammkapital = code[7].text
        else:
            stammkapital = ""
    else:
            stammkapital = ""
    
    data = {}
    data['Address'] = name , address,postal_code,addressLocality
    data['Telephone'] = telephone
    data['Fax_number'] = fax_number
    data['Email'] = email
    data['Internet'] = internet
    data['Ofizieller Name'] = ofizieller
    data['Hauptsitz'] = hauptsitz
    data['Branche'] = branche
    data['Unternehmensgegenstand'] = unternehmensgegenstand
    data['Handelsregister'] = handelsregister
    data['Rechtsform'] = rechtsform
    data['Firmengründung'] = firmengründung
    data['Stammkapital'] = stammkapital
    
    company_data.append(data)

    print(data)

 
```
